### Pyplot API vs Object-Oriented API
- PA -> Quickly
- OO API -> Advanced

```python
x = np.linspace(0,10,1000);
plt.plot(x,np.sin(x), color = "blue", linestyle = "dashed", label = "Sin x");
plt.plot(x,np.cos(x), color = "red", label = "Cosin x");
plt.title("Sin(x) and Cos(x)")
plt.xlabel("Bien x");
plt.ylabel("Sin x")
#plt.xlim([0,4])
plt.legend();
```

Object-Oriented API

```python
x= [1,2,3,4]
y = [11,22,33,44]
fig, ax=plt.subplots()
ax.plot(x,y)
```

```python
rng = np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
sizes = 1000*rng.randn(100)
color = rng.randn(100)
fig, ax=plt.subplots()
im1=ax.scatter(x,y, s=sizes, c=color, cmap = "viridis", alpha=0.3);
fig.colorbar(im1)
```

### 3. Bar chart
* Horizontal
* Vertical

```python
soft_drink_price={"coke":10, "pepsi":12,"sprite":8}
fig, ax = plt.subplots();
ax.bar(soft_drink_price.keys(),soft_drink_price.values());
ax.set(title = "Bach hoa xanh", ylabel = "Price", xlabel = "Do uong");
```

```python
fig, ax = plt.subplots();
ax.barh(list(soft_drink_price.keys()),list(soft_drink_price.values()));
```

### 4. Histogram

```python
np.random.seed(42)
student_height = np.random.normal(170,10,250)
```

```python
student_height[:10]
```

```python
plt.hist(student_height, bins=30);
```

```python
# Option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_price.keys(), soft_drink_price.values());
ax4.hist(student_height, bins=30);
```
